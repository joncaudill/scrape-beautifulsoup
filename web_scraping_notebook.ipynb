```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
wiki_url = "https://en.wikipedia.org/wiki/List_of_Disney_theme_park_attractions"
wiki_data = urlopen(wiki_url)
wiki_html = wiki_data.read()
wiki_data.close()

page_soup = soup(wiki_html, "html.parser")
```

```python
print(page_soup)
```

```python
print(page_soup.h1)
```

```python
attractions_table = page_soup.find_all('table', {'class': 'wikitable sortable'})
print(attractions_table)
```

```python
attractions_table = attractions_table[0]
headers = attractions_table.find_all('th', {})
print(headers)
```

```python
header_titles = []
for header in headers:
    header_titles.append(header.get_text(strip=True))
print(header_titles)
```

```python
all_rows = attractions_table.find_all('tr', {})
print(all_rows)
```

```python
data = all_rows[2:]
print(data)
```

```python
first_row = data[0]
first_row_data = first_row.find_all('td', {})
print(first_row_data)
```

```python
data_texts = []
for data_text in first_row_data:
    data_texts.append(data_text.get_text(strip=True))
print(data_texts)
```

```python
table_rows = []
for row in data:
    row_data = row.find_all('td', {})
    row_texts = []
    for data_text in row_data:
        data_cleaned = data_text.get_text(strip=True)
        data_cleaned = data_cleaned.replace(',', '')
        row_texts.append(data_cleaned)
    table_rows.append(row_texts)
print(table_rows)
```

```python
filename = "disney_attractions.csv"
with open(filename, "w") as file:
    file.write(",".join(header_titles) + "\n")
    for row in table_rows:
        file.write(",".join(row) + "\n")
print(f"Data has been written to {filename}")
```
